```python
def entferne_ausreisser_iqr(df, spalten):
    for spalte in spalten:
        Q1 = df[spalte].quantile(0.25)
        Q3 = df[spalte].quantile(0.75)
        IQR = Q3 - Q1
        untere_grenze = Q1 - 1.5 * IQR
        obere_grenze = Q3 + 1.5 * IQR
        df = df[(df[spalte] >= untere_grenze) & (df[spalte] <= obere_grenze)]
    return df.reset_index(drop=True)

data_clean = entferne_ausreisser_iqr(data, ausseit)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def zeige_spaltenverteilungen(df, max_kategorien=20, bins=30):
    farben_fest = [color_compare1, color_compare2, color_main]
    
    for i, spalte in enumerate(df.columns):
        daten = df[spalte].dropna()
        
        if daten.empty:
            continue
            
        if pd.api.types.is_numeric_dtype(daten):
            fig, ax = plt.subplots(figsize=(7, 4))
            fig.patch.set_facecolor('none')
            ax.set_facecolor('none')
            
            counts, edges = np.histogram(daten, bins=bins)
            prozente = 100 * counts / counts.sum()
            colors = [farben_fest[j % len(farben_fest)] for j in range(len(prozente))]
            
            ax.bar(edges[:-1], prozente, width=np.diff(edges),
                   edgecolor="black", align="edge", color=colors)
            ax.set_ylabel("Relative Häufigkeit (%)")
            ax.set_xlabel(spalte)
            ax.set_title(f'Histogramm: {spalte}')
            
        else:
            werte = daten.value_counts(normalize=True) * 100
            if len(werte) > max_kategorien:
                werte = werte.iloc[:max_kategorien]
            
            fig, ax = plt.subplots(figsize=(10, 5))  # Nur hier Figure erstellen
            fig.patch.set_facecolor('none')
            ax.set_facecolor('none')
            
            x = np.arange(len(werte))
            colors = [farben_fest[i % len(farben_fest)] for i in range(len(werte))]
            
            ax.bar(x, werte.values, color=colors, edgecolor='black')
            ax.set_xticks(x)
            ax.set_xticklabels(werte.index.astype(str), rotation=60, ha='right')
            ax.set_ylabel("Relative Häufigkeit (%)")
            ax.set_xlabel(spalte)
            ax.set_title(f"Balkendiagramm: {spalte}")
        
        ax.grid(True, linestyle="--", alpha=0.5)
        plt.tight_layout()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import to_rgb

farben = {
    'W': color_main, #von compare farben auf main gewechselt
    'M': color_main
}

farben_final = {
    'M_über_18': farben['M'],
    'M_unter_18': farben['M'],
    'W_über_18': farben['W'],
    'W_unter_18': farben['W']
}

# Altersgruppen definieren
data_clean['alter_gruppe'] = data_clean["Alter Einstufung"].apply(
    lambda x: 'unter_18' if x < 18 else 'über_18'
)

# Globale Min/Max für Einstufungswert zuletzt
daten_gesamt = data_clean["Einstufungswert zuletzt"].dropna()
xmin, xmax = daten_gesamt.min(), daten_gesamt.max()

# Bins mit 50er Schritten, gerundet nach unten und oben
bins = np.arange((xmin // 50) * 50, ((xmax // 50) + 2) * 50, 50)

fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
fig.patch.set_facecolor('none')

for i, geschlecht in enumerate(['M', 'W']):
    for j, altersgruppe in enumerate(['unter_18', 'über_18']):
        subset = data_clean[
            (data_clean["GeschlechtImSpielbetrieb"] == geschlecht) &
            (data_clean['alter_gruppe'] == altersgruppe)
        ]
        daten = subset["Einstufungswert zuletzt"].dropna()
        n = len(daten)

        if n == 0:
            continue

        counts, edges = np.histogram(daten, bins=bins)
        prozent = counts / n * 100
        mitte = 0.5 * (edges[1:] + edges[:-1])

        ax = axs[i, j]
        ax.bar(mitte, prozent, width=(edges[1] - edges[0]),
               color=farben_final[f"{geschlecht}_{altersgruppe}"], edgecolor='black')

        titel = f"{'Männlich' if geschlecht == 'M' else 'Weiblich'} – {altersgruppe.replace('_', ' ')} (n = {n})"
        ax.set_title(titel, color='black', fontsize=12)

        ax.set_facecolor('none')
        ax.tick_params(colors='black')
        ax.grid(True, color='gray', linestyle='--', alpha=0.5)
        if i == 1:
            ax.set_xlabel("TTR-Wert", color='black')
        if j == 0:
            ax.set_ylabel("Häufigkeit [%]", color='black')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Filter: Männer ≥18 mit TTR-Werten
maske = (
    (data_clean["GeschlechtImSpielbetrieb"] == "M") &
    (data_clean["Alter Einstufung"] >= 18) &
    (data_clean["Einstufungswert zuletzt"].notna()) &
    (data_clean["TTR_Q4"].notna())
)

df = data_clean.loc[maske].copy()

# 2. Differenzspalte
df["Delta_Q4"] = df["TTR_Q4"] - df["Einstufungswert zuletzt"]

# 3. TTR-Klassen definieren (falls noch nicht geschehen)
df["TTR_Klasse"] = pd.cut(
    df["Einstufungswert zuletzt"],
    bins=[900, 1100, 1300, 1500],
    labels=["900–1100", "1100–1300", "1300–1500"],
    right=False
)

# 4. Plot-Setup
klassen = ["900–1100", "1100–1300", "1300–1500"]
farben = {
    "900–1100": color_compare1,
    "1100–1300": color_main,
    "1300–1500": color_compare2
}

# Einheitliches Binning
bin_edges = np.arange(-300, 310, 20)  # von -300 bis +300 in 20er-Schritten
bin_mitte = 0.5 * (bin_edges[:-1] + bin_edges[1:])
bin_width = bin_edges[1] - bin_edges[0]

fig, axs = plt.subplots(1, 3, figsize=(16, 5), sharey=True)
fig.patch.set_facecolor('none')

for i, klasse in enumerate(klassen):
    subset = df[df["TTR_Klasse"] == klasse]
    deltas = subset["Delta_Q4"].dropna()
    n = len(deltas)

    if n == 0:
        continue

    # Histogramm-Daten berechnen
    counts, _ = np.histogram(deltas, bins=bin_edges)
    prozent = counts / n * 100

    # Median
    median = deltas.median()

    # Plotten
    ax = axs[i]
    ax.bar(bin_mitte, prozent, width=bin_width, color=farben[klasse], edgecolor='black')

    # Rote Medianlinie
    ax.axvline(median, color='red', linestyle='--', linewidth=2, label=f'Median: {median:.1f}')

    # Layout
    ax.set_title(f"TTR-Klasse {klasse} (n = {n})", color='black')
    ax.set_xlabel("Δ Q4 – Einstufung", color='black')
    if i == 0:
        ax.set_ylabel("Häufigkeit [%]", color='black')

    ax.legend()
    ax.set_facecolor('none')
    ax.tick_params(colors='black')
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.set_xlim(-300, 300)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Daten filtern
maske = (
    data_clean[["TTR_Q4", "Einstufungswert zuletzt", "Einstufungsart zuletzt"]].notna().all(axis=1)
)

df = data_clean.loc[maske].copy()

# 2. Differenz berechnen
df["Delta_Q4"] = df["TTR_Q4"] - df["Einstufungswert zuletzt"]

# 3. Gruppierung
gruppenname = "Einstufungsart zuletzt"
gruppen = ["P", "T"]
labels_dict = {"P": "Punktspielinitialisierung", "T": "Turnierinitialisierung"}
farben = {"P": color_compare1, "T": color_compare2}

# 4. Histogramm-Einstellungen
bin_edges = np.arange(-300, 310, 20)
bin_mitte = 0.5 * (bin_edges[:-1] + bin_edges[1:])
bin_width = bin_edges[1] - bin_edges[0]

# 5. Histogramm-Werte berechnen + Maxima für Skalierung
prozent_daten = {}
max_y = 0
for gruppe in gruppen:
    deltas = df.loc[df[gruppenname] == gruppe, "Delta_Q4"].dropna()
    n = len(deltas)
    counts, _ = np.histogram(deltas, bins=bin_edges)
    prozent = counts / n * 100 if n > 0 else np.zeros_like(counts)
    prozent_daten[gruppe] = (prozent, n)
    max_y = max(max_y, prozent.max())

# 6. Layout: 2 Zeilen, 2 Spalten (60/40 Verhältnis)
fig = plt.figure(figsize=(16, 8))
gs = fig.add_gridspec(2, 2, width_ratios=[5, 5], height_ratios=[1, 1], wspace=0.3, hspace=0.3)
fig.patch.set_facecolor('none')

ax_p = fig.add_subplot(gs[0, 0])
ax_t = fig.add_subplot(gs[1, 0])
ax_box = fig.add_subplot(gs[:, 1])

# 7. Barplots zeichnen (links)
for gruppe, ax in zip(gruppen, [ax_p, ax_t]):
    prozent, n = prozent_daten[gruppe]
    ax.bar(bin_mitte, prozent, width=bin_width, color=farben[gruppe], edgecolor='black')
    ax.set_title(f"{labels_dict[gruppe]} (n={n})", color='black')
    ax.set_xlabel("TTR Veränderung", color='black')
    ax.set_ylabel("Häufigkeit [%]", color='black')
    ax.set_xlim(-300, 300)
    ax.set_ylim(0, max_y * 1.2)  # Einheitliche Skalierung
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.set_facecolor('none')
    ax.tick_params(colors='black')

# 8. Boxplot (rechts)
daten = [df.loc[df[gruppenname] == g, "Delta_Q4"].dropna() for g in gruppen]
box = ax_box.boxplot(
    daten,
    patch_artist=True,
    labels=[labels_dict[g] for g in gruppen],
    widths=0.4,  # Etwas schmaler
    medianprops=dict(color='purple', linewidth=2),
    boxprops=dict(linewidth=1.5),
    whiskerprops=dict(linewidth=1.2),
    capprops=dict(linewidth=1.2),
    flierprops=dict(marker='o', markersize=4, alpha=0.5)
)

# Farben für Boxplots
for patch, gruppe in zip(box['boxes'], gruppen):
    patch.set_facecolor(farben[gruppe])
    patch.set_edgecolor('black')

ax_box.set_title("Verteilung der Initialisierungsverfahren", color='black')
ax_box.set_ylabel("TTR-Veränderung", color='black')
ax_box.set_yticks(np.arange(-300, 301, 25))
ax_box.set_ylim(-300, 300)
ax_box.grid(True, linestyle='--', alpha=0.4)
ax_box.set_facecolor('none')
ax_box.tick_params(colors='black')

# 2. Beschriftungen nur alle 100 Punkte
tick_labels = []
for val in np.arange(-300, 301, 25):
    if val % 75 == 0:
        tick_labels.append(str(val))
    else:
        tick_labels.append("")  # keine Zahl anzeigen

ax_box.set_yticklabels(tick_labels)



plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Filter für gültige Daten und relevante Spalten
maske = (
    data_clean[["TTR_Q4", "Einstufungswert zuletzt", "Einzel_Q1", "Einzel_Q2", "Einzel_Q3", "Einzel_Q4"]]
    .notna()
    .all(axis=1)
)
df = data_clean.loc[maske].copy()

# Berechnung der TTR-Veränderung Q4 - Einstufungswert zuletzt
df["Delta_Q4"] = df["TTR_Q4"] - df["Einstufungswert zuletzt"]

# Kumulative gespielte Einzelspiele bis Q4
df["Einzel_Q_gespielt"] = df["Einzel_Q1"] + df["Einzel_Q2"] + df["Einzel_Q3"] + df["Einzel_Q4"]

# Bins für gespielte Einzelspiele (angepasst an das Zielbild)
bin_edges = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
bin_centers = [(bin_edges[i] + bin_edges[i+1]) / 2 for i in range(len(bin_edges)-1)]
bin_labels = [f"[{bin_edges[i]}, {bin_edges[i+1]})" for i in range(len(bin_edges)-1)]

df["Spielgruppen"] = pd.cut(df["Einzel_Q_gespielt"], bins=bin_edges, labels=bin_labels, right=False)

# Berechnung Mittelwerte und Standardfehler je Gruppe
stats = df.groupby("Spielgruppen")["Delta_Q4"].agg(['mean', 'std', 'count']).reset_index()

# Standardfehler des Mittelwerts (SE)
stats['se'] = stats['std'] / np.sqrt(stats['count'])

# Nur Gruppen mit ausreichend Daten behalten
stats = stats[stats['count'] >= 5].reset_index(drop=True)

# Entsprechende bin_centers anpassen
valid_bin_centers = [bin_centers[i] for i in range(len(bin_labels)) 
                    if bin_labels[i] in stats["Spielgruppen"].values]

# Visualisierung - nur ein Plot
plt.figure(figsize=(10, 6))


# Liniendiagramm mit Fehlerbalken
plt.errorbar(valid_bin_centers, stats["mean"], yerr=stats["se"], 
             marker='o', color=color_main, linewidth=2, markersize=6, 
             capsize=5, capthick=2, elinewidth=2)

plt.title("TTR-Veränderung nach gespielten Einzelspielen", 
          fontsize=14, pad=20)
plt.xlabel("Gespielte Einzelspiele (Intervalle)", fontsize=12)
plt.ylabel("Durchschnittliche TTR-Veränderung", fontsize=12)

# Grid wie im Zielbild
plt.grid(True, linestyle='-', alpha=0.3, color='gray')

# Achsenbeschriftung anpassen
plt.xticks(valid_bin_centers, 
           [f"[{int(center-25)}, {int(center+25)})" for center in valid_bin_centers], 
           rotation=45)

# Y-Achse anpassen
y_min = min(stats["mean"] - stats["se"]) * 1.1
y_max = max(stats["mean"] + stats["se"]) * 1.1
plt.ylim(y_min, y_max)


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Filter: Männer ≥18 mit gültigen Werten in Q0-Q4, Einstufungswert zuletzt und Einstufungsart
maske = (
    (data_clean["GeschlechtImSpielbetrieb"] == "M") &
    (data_clean["Alter Einstufung"] >= 18) &
    data_clean[["TTR_Q0", "TTR_Q1", "TTR_Q2", "TTR_Q3", "TTR_Q4", "Einstufungswert zuletzt"]].notna().all(axis=1) &
    (data_clean["Einstufungsart zuletzt"].notna())
)

df = data_clean.loc[maske].copy()

# 2. TTR-Klassen definieren
df["TTR_Klasse"] = pd.cut(
    df["Einstufungswert zuletzt"],
    bins=[900, 1100, 1300, 1500],
    labels=["900–1100", "1100–1300", "1300–1500"],
    right=False
)

# 3. Einstufungsarten filtern (nur P und T)
arten = [a for a in ['P', 'T'] if a in df["Einstufungsart zuletzt"].unique()]

quartale = ["TTR_Q0", "TTR_Q1", "TTR_Q2", "TTR_Q3", "TTR_Q4"]
quartal_labels = ["Q0", "Q1", "Q2", "Q3", "Q4"]

fig, axs = plt.subplots(len(arten), 3, figsize=(18, 10), sharey=True)
fig.patch.set_facecolor('none')

for i, art in enumerate(arten):
    for j, klasse in enumerate(["900–1100", "1100–1300", "1300–1500"]):
        ax = axs[i, j] if len(arten) > 1 else axs[j]

        subset = df[(df["Einstufungsart zuletzt"] == art) & (df["TTR_Klasse"] == klasse)]

        if subset.empty:
            ax.set_visible(False)
            continue

        # 4. Differenzen pro Quartal berechnen: DataFrame mit Spalten Q0..Q4 als Differenz zum Einstufungswert
        delta_df = pd.DataFrame({
            q: subset[q] - subset["Einstufungswert zuletzt"]
            for q in quartale
        })

        # 5. Daten für Boxplot: Liste von Series (pro Quartal)
        daten_box = [delta_df[q].dropna() for q in quartale]

        ax.boxplot(daten_box, labels=quartal_labels, patch_artist=True,
                   boxprops=dict(facecolor='#ff6a00', color='black'),
                   medianprops=dict(color='purple', linewidth=2))

        ax.set_title(f"{'Punktspiel' if art == 'P' else 'Turnier'} – {klasse} (n={len(subset)})", color='black')
        ax.set_xlabel("Quartal", color='black')
        if j == 0:
            ax.set_ylabel("TTR Veränderung", color='black')

        ax.tick_params(colors='black')
        ax.grid(True, linestyle='--', alpha=0.4)
        ax.set_facecolor('none')
        ax.set_ylim(-300, 300)  # je nach sinnvoller Range anpassen

plt.tight_layout()
plt.show()
```
