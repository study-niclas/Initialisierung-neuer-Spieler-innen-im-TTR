```python
import duckdb
import pandas as pd 

con = duckdb.connect("qttr.db")

tabellen = con.execute("SHOW TABLES").fetchdf()["name"].tolist()
print(tabellen)
qttr2025 = con.execute(f'SELECT * FROM "{tabellen[-1]}" ').fetchdf()
qttr2025
```

```python
df = con.execute(f'''
    SELECT * 
    FROM "{tabellen[-1]}"
    WHERE TRY_CAST("Einstufungsdatum zuletzt" AS DATE)  BETWEEN DATE '2020-01-01' AND DATE '2024-02-10'
    AND "Kumulierte Inaktivitätsabzüge" = 0
    AND "Status" = 'a'
''').fetchdf()
df
```

```python
df.columns.tolist()
```

```python
relevante_spalten = ['InterneNr','Nachname','Vorname','GeschlechtImSpielbetrieb','Geburtsdatum','Einstufungsdatum zuletzt','TTR','Anzahl Einzel gesamt','Einstufungsart zuletzt','Nation','Altersklasse','Spielklasse','Einstufungswert zuletzt']
relevante_spalten
```

```python
df = df[relevante_spalten]
df.sort_values('TTR')
```

```python
import re
from datetime import datetime

# nur relevante Tabellen filtern + Datum extrahieren
quartalstabellen = []
for name in tabellen:
    match = re.search(r"q-ttr-liste_(\d{8})_komplett", name)
    if match:
        datum = datetime.strptime(match.group(1), "%Y%m%d")
        if datum.year >= 2020:
            quartalstabellen.append((datum, name))


# chronologisch sortieren
quartalstabellen.sort()
quartalstabellen
```

```python
df["Einstufungsdatum zuletzt"] = pd.to_datetime(df["Einstufungsdatum zuletzt"], format="%d.%m.%Y", errors="coerce")
```

```python
# leeres Ergebnis-DataFrame, beginnt mit df_neu
result_df = df.copy()

# Alle Spieler mit Einstufungsdatum
for idx, row in df.iterrows():
    interne_nr = row["InterneNr"]
    einstufung = row["Einstufungsdatum zuletzt"]

    # Finde Index der Quartalstabelle, die >= Einstufungsdatum ist
    quartale = [d for (d, _) in quartalstabellen]
    index_q0 = next((i for i, d in enumerate(quartale) if d >= einstufung), None)
    
    if index_q0 is None or index_q0 + 4 >= len(quartalstabellen):
        continue  # überspringen, wenn nicht genug Quartale vorhanden sind
    
    # Sammle Werte aus Q0 bis Q4
    for offset in range(5):
        datum, tabname = quartalstabellen[index_q0 + offset]
        quartal_label = f"Q{offset}"
        
        row_data = con.execute(f'''
            SELECT "TTR", "Anzahl Einzel gesamt"
            FROM "{tabname}"
            WHERE "InterneNr" = '{interne_nr}'
        ''').fetchdf()
        
        if not row_data.empty:
            result_df.loc[idx, f"TTR_{quartal_label}"] = row_data.at[0, "TTR"]
            result_df.loc[idx, f"Einzel_{quartal_label}"] = row_data.at[0, "Anzahl Einzel gesamt"]
```
