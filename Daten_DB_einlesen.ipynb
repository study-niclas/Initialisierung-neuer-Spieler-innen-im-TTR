```python
import duckdb
import pandas as pd
import os

# üìÅ Ordner mit deinen Excel-Dateien
folder = 
file_paths = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith((".xlsx", ".xls"))]

# üíæ DuckDB-Verbindung
con = duckdb.connect("qttr.db")

# üìä Alle Dateien in Tabellen schreiben
for path in file_paths:
    table_name = os.path.splitext(os.path.basename(path))[0].lower()
    df = pd.read_excel(path).astype(str)
    con.register("temp_df", df)
    con.execute(f"CREATE OR REPLACE TABLE '{table_name}' AS SELECT * FROM temp_df")
    con.unregister("temp_df")

print("Alle Excel-Dateien aus dem Ordner wurden in DuckDB gespeichert.")
```

```python
import duckdb
import pandas as pd

con = duckdb.connect("qttr.db")

tabellen = con.execute("SHOW TABLES").fetchdf()["name"].tolist()
tabellen
```

```python
#spalten = ["Rang",'Deutscher Rang', 'Einstufungswert zuletzt', 'TTR','Anzahl Einzel gesamt','Kumulierte Inaktivit√§tsabz√ºge','Kumulierte QNSZ-Werte','Kumulierte QNEB-Werte', '2J','lebt','MM','spb','10E','Typ']  # Liste der Spalten, die umgewandelt werden sollen
spalten = ["Einstufungsdatum zuletzt"]

for tabelle in tabellen:
    spalten_liste = con.execute(f"PRAGMA table_info('{tabelle}')").fetchdf()["name"].tolist()

    for spalte in spalten:
        if spalte in spalten_liste:
            
            con.execute(f"""
                CREATE OR REPLACE TABLE "{tabelle}" AS
                SELECT 
                    *, 
                    TRY_CAST("{spalte}" AS DATE) AS "{spalte}_neu"
                FROM "{tabelle}"
            """) 
            
            con.execute(f"""ALTER TABLE "{tabelle}" DROP COLUMN "{spalte}";""")
            con.execute(f"""ALTER TABLE "{tabelle}" RENAME COLUMN "{spalte}_neu" TO "{spalte}";""")
```
